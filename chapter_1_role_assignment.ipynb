- Inside resource group , access control(IAM) -> main three roles : owner, contributor, reader
- Owner : full access to all resources, can delegate access to others
- Contributor : can create and manage all types of Azure resources but can't grant access to others
- Reader : can view existing Azure resources but can't make any changes

- in real time it will be difficult to assign role to each user, so we can create groups and assign role to group and add users to group
