## in adf there are 5 types of pages:
- home
- auuthor
- Monitor
- manage
- learning center

## Three important things to connect to data ssocurces in adf:
- under Manage, you can create ""linked services"", which are connections to data sources.
- under Manage, you can also create ""integration runtimes"", which are compute resources that are used to run pipelines.
- under Author, you can create ""datasets"", which are representations of data that you want to use in your pipelines.

## Home page:
- here you can see the pipelines that you have recently worked on, and you can also create new pipelines from here.
- you can also see the recent activities that you have done in adf, and you can also see the recent pipelines that you have run.
- you can also see the recent datasets that you have created, and you can also see the recent linked services that you have created.
