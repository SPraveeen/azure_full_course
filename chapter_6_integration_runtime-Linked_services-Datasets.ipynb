## integration runtimes:
- three types of integration runtimes:
    - azure integration runtime
    - self-hosted integration runtime
    - SSIS integration runtime
- azure integration runtime: this is a fully managed, serverless compute resource that is used to run pipelines in Azure Data Factory. It is the default integration runtime that is used when you create a new pipeline.
- self-hosted integration runtime: this is a compute resource that you can install on-premises or in a virtual machine in Azure. It is used to run pipelines that need to access data sources that are not accessible from the Azure integration runtime.
- SSIS integration runtime: this is a compute resource that is used to run SQL Server Integration Services (SSIS) packages in Azure Data Factory. It is used to run pipelines that need to access data sources that are not accessible from the Azure integration runtime, and that require the use of SSIS packages.

# Linked services:
- linked services are connections to data sources. They are used to connect to data sources that are not accessible from the Azure integration runtime. They are also used to connect to data sources that require authentication.
